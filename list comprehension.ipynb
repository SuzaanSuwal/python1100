````python
nums = {8, 3, 4, 5}
sq = []
for item in nums:
    sq.append(item** 2)
sq
````

````python
nums = {8, 3, 4, 5}
sq = [item ** 2 for item in nums]
sq
````

# List comprehension with if else
```python
[Expression if True else expresion for item in iterable]
```
https://www.nomidl.com/wp-content/uploads/2022/04/image-10.png


````python
x = range(-10, 11)
````

````python
output = []
for item in x:# condition
    if item <= 0: # if ko expression
        output.append(0)
    else: # else ko expression
        output.append(item)
output
````

````python
output = [0 if item <= 0 else item for item in x]
output
````
